Weâ€™ll use Python libraries: `numpy`, `scipy.stats`, and `matplotlib` for computations and visualizations.

### ðŸ”¹ Mean
The **mean** is the average of all values in a dataset.  

### ðŸ”¹ Median
The **median** is the middle value when data is sorted.  

### ðŸ”¹ Variance
The **variance** measures the spread of data from the mean.  

### ðŸ”¹ Standard Deviation
The **standard deviation** is the square root of variance, showing how spread out data is.


```python
import numpy as np

# Sample dataset
data = [10, 15, 20, 25, 30]

# Manual functions
def mean(lst):
    return sum(lst)/len(lst)

def variance(lst):
    mu = mean(lst)
    return sum((x - mu)**2 for x in lst)/len(lst)

def std_dev(lst):
    return variance(lst)**0.5

print("Data:", data)
print("Mean:", mean(data))
print("Median:", np.median(data))
print("Variance:", variance(data))
print("Standard Deviation:", std_dev(data))
```

## Probability Distributions

A probability distribution describes how values are distributed.

- **Normal Distribution:** Bell-shaped curve.
- **Binomial Distribution:** Number of successes in fixed trials.
- **Poisson Distribution:** Number of events in fixed interval.

```python

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, binom, poisson

# Normal Distribution
samples_normal = np.random.normal(loc=0, scale=1, size=1000)
sns.histplot(samples_normal, kde=True, bins=30)
plt.title("Normal Distribution (mean=0, std=1)")
plt.show()

# Binomial Distribution - Dice rolls
n, p = 10, 0.5  # 10 trials, prob success=0.5
x = np.arange(0, n+1)
pmf_binom = binom.pmf(x, n, p)
plt.bar(x, pmf_binom)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()

# Poisson Distribution
mu = 3
x = np.arange(0, 15)
pmf_poisson = poisson.pmf(x, mu)
plt.bar(x, pmf_poisson)
plt.title("Poisson Distribution (mu=3)")
plt.show()
```

## Bayes' Theorem

Bayesâ€™ theorem describes the probability of an event based on prior knowledge.

**Formula:**

$$ P(A|B) = \frac{P(B|A) P(A)}{P(B)} $$

Example: Probability a student knows the answer given they guessed correctly.

```python

# Example: Student answering a question
# P(K) = probability student knows the answer = 0.6
# P(G) = probability student guesses = 0.4
# If they know: P(Correct|K) = 1
# If they guess: P(Correct|G) = 0.25

P_K = 0.6
P_G = 0.4
P_Correct_given_K = 1
P_Correct_given_G = 0.25

# Total probability of correct answer
P_Correct = (P_Correct_given_K * P_K) + (P_Correct_given_G * P_G)

# Bayes: P(K|Correct)
P_K_given_Correct = (P_Correct_given_K * P_K) / P_Correct
print("Probability student knows the answer given they answered correctly:", P_K_given_Correct)
```

### 4.1 Coin Toss Simulation
Simulate tossing a fair coin 1000 times and calculate the probability of getting heads.

```python

import random

trials = 1000
heads = sum(1 for _ in range(trials) if random.choice(['H','T']) == 'H')
prob_heads = heads / trials
print(f"Heads count: {heads}, Probability of Heads: {prob_heads}")
```
